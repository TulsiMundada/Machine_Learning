```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Users\dai\Desktop\Machine Learning\Datasets\monthly-milk-production-pounds-p.csv",index_col=0)
df1 = pd.read_csv("C:/Users\dai\Desktop\Machine Learning\Datasets\monthly-beer-production-in-austr.csv",index_col=0)
```

```python
df.index = pd.to_datetime(df.index).to_period("M")
```

```python
df.plot()
plt.title("Monthly Milk Production")
plt.xlabel("Months")
plt.show()
```

```python
downsampled = df.resample("Q").sum()
downsampled.index.rename("Quarter",inplace=True)
downsampled.plot()
plt.title("Quarterly milk production")
plt.xlabel("Quarters")
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
df = pd.read_csv("C:/Users\dai\Desktop\Machine Learning\Datasets\monthly-milk-production-pounds-p.csv")

series = df['Milk']
result = seasonal_decompose(series,model='additive',period =12)
result.plot()
plt.title("Additive Decompositon")
plt.show()
```

```python
result = seasonal_decompose(series,model='multiplicative',period =12)
result.plot()
plt.title("Multiplicative Decompositon")
plt.show()
```

```python
df1.index = pd.to_datetime(df1.index).to_period("M")
df1.plot()
plt.title("Monthly Beer Production")
plt.xlabel("Months")
plt.show()
downsampled = df1.resample("Q").sum()
downsampled.index.rename("Quarter",inplace=True)
downsampled.plot()
plt.title("Quarterly Beer production")
plt.ylabel("Quarters")
plt.show()
```
