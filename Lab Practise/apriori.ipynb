Association Rule

```python
import matplotlib.pyplot as plt
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import os
```

```python
os.chdir("C:/Users/dai/Desktop/Machine Learning/Datasets")

fp_df = pd.read_csv("Faceplate.csv",index_col=0)
fp_df.head()
```

```python
fp_df = fp_df.astype(bool)
itemsets = apriori(fp_df , min_support=0.2, use_colnames=True)

rules = association_rules(itemsets, metric='confidence',min_threshold=0.6)
print(rules[['antecedents','consequents','support','confidence','lift']])
```

```python
rules.sort_values(by=['lift'],ascending=False).head(6)
```

```python
rules_df = rules.sort_values(by=['lift','confidence'],ascending=False)
print(rules_df[['antecedents','consequents','support','confidence','lift']])
```

```python
rely_rules = rules_df[rules_df['lift']>1]
print(rely_rules[['antecedents','consequents','support','confidence','lift']])
```
