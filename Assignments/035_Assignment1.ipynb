```python
os.chdir(r"C:\Users\dai\Desktop\Machine Learning\Assignments\Assignment 1")
test = pd.read_csv("test.csv")
train = pd.read_csv("train.csv",index_col=0)
```

```python
''' Try the following models:
Logistic Regression
Random Forest
X G Boost
Light GBM
Stacking with models of your choice

Mention leaderboard scores for each of the five.
'''
```

```python
le = LabelEncoder()
y_train = train['Rings']
s_en=pd.DataFrame(le.fit_transform(train['Sex']))
X_train= train.drop('Sex',axis=1)
X_train['Sex']=s_en
X_train=X_train.drop('Rings',axis=1)
X_train
```

```python
s_en=pd.DataFrame(le.fit_transform(test['Sex']))
X_test= test.drop('Sex',axis=1)
X_test['Sex']=s_en
X_test=X_test.drop('id',axis=1)
X_test
```

```python
#using lgb
```

```python
lgb=LGBMRegressor()
```

```python
lgb.fit(X_train,y_train)
```

```python
y_pred=lgb.predict(X_test)
```

```python
submit=pd.DataFrame({'id':test['id'],'Rings':y_pred})
```

```python
submit.to_csv('lgb_abalone.csv',index=False)
```
