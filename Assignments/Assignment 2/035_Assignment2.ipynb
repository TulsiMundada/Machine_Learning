```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.metrics import roc_curve, roc_auc_score
from sklearn.metrics import log_loss
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.model_selection import StratifiedKFold, KFold
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
import os
```

```python

os.chdir(r"C:\Users\dai\Desktop\Machine Learning\Assignments\Assignment 2\santander-customer-satisfaction")
test = pd.read_csv("test.csv")
train = pd.read_csv("train.csv",index_col=0)
```

```python
X_train = train.drop(['TARGET'], axis=1)
y_train = train['TARGET']
X_test =  test.drop('ID', axis=1)
print(X_train.columns)
print(X_test.columns)
```

```python
#Find out how many principal components explain more than 90% variation 
#taking all the variables except ID and target. 
```

```python
scaler = StandardScaler().set_output(transform='pandas')
scaler
```

```python
scaler.fit(train)
```

```python
train_scl = scaler.transform(train)
```

```python
pca = PCA().set_output(transform='pandas')
principalComponents = pca.fit_transform(train_scl)
```

```python
print(principalComponents.var())
```

```python
print(np.cumsum(pca.explained_variance_ratio_*100))
```

```python
var=(np.cumsum(pca.explained_variance_ratio_*100))
```

```python
count=(var > 90).sum()
print("no of features that give varaiation more than 90% is: ",count)
```

```python
scaler = StandardScaler().set_output(transform='pandas')
prcomp=PCA().set_output(transform='pandas')
```

```python
#using random forest
```

```python
rf=RandomForestClassifier(random_state=24)
```

```python
pipe_train=Pipeline([('SCL',scaler),('PCA',prcomp),('RF',rf)])
pipe_train.fit(X_train,y_train)
```

```python
y_pred=pipe_train.predict(X_test)
```

```python
y_pred
```

```python
submit=pd.DataFrame({'ID':test['ID'],'TARGET':y_pred})
```
